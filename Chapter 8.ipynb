```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'key':['b','b','a','c','a','a','b'],'data1':range(7)})
df1
```

```python
df2=pd.DataFrame({'key':['a','b','a','d'],'data2':range(4)})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df2,df1)
```

```python
pd.merge(df1,df2, on='key')
```

```python
df3=pd.DataFrame({'left_key':['b','b','a','c','a','a','b'],'data1':range(7)})
```

```python
df4=pd.DataFrame({'right_key':['a','b','a','d'],'data2':range(4)})
```

```python
pd.merge(df3,df4, left_on='left_key', right_on='right_key')
```

```python
pd.merge(df3,df4, left_on='left_key', right_on='right_key', how='outer')
```

```python
pd.merge(df3,df4, left_on='left_key', right_on='right_key', how='outer')
```

```python
pd.merge(df1,df2, on='key', how='left')
```

```python
pd.merge(df1,df2, on='key', how='right')
```

```python
dict={'userID':['foo','foo','bar'], 'userName':['one','two','one'], 'lval':[1,2,3]}
left=pd.DataFrame(dict)
left
```

```python
dict1={'userID':['foo','foo','bar','bar'], 'userName':['one','one','one','two'], 'rval':[4,5,6,7]}
right=pd.DataFrame(dict1)
right
```

```python
pd.merge(left,right, on=['userID', 'userName'], how='right')
```

```python
pd.merge(left,right, on=['userID', 'userName'], how='left')
```

```python
pd.merge(right,left, on=['userID', 'userName'], how='right')
```

```python
pd.merge(right,left, on=['userID', 'userName'], how='left')
```

```python
left1=pd.DataFrame({'key':['c','d','c','d','a','b'],'value': range(6)})
left1
```

```python
right1=pd.DataFrame({'group_val':[3,5,7]})
right1
```

```python
pd.merge(left1,right1, left_index=True, right_index=True)
```

```python
left1.join(right1, how='inner')
```
